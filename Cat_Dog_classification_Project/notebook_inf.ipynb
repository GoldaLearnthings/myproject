## import library

```python
# Import libraries
import pickle
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image
```

## load model

```python
CNNSequential = load_model('model_1.h5')
```

## import gambar

```python
!curl https://www.kucingklik.com/wp-content/uploads/2019/02/kucing.jpg --output predict.jpg
```

```python
img = image.load_img('predict.jpg')
img
```

```python
img = image.img_to_array(img) / 220
img = tf.image.resize(img, (220,220))
img = tf.expand_dims(img, axis=0)
```

mengubah ukuran gambar 220 * 220

## predict

```python
prediction = int(tf.round(CNNSequential.predict(x = img)).numpy()[0][0])

def classpredict(value):
    if value == 0:
        return "cat"
    elif value == 1:
        return "dog"

print('Image 1 Prediction : ', classpredict(prediction))
```

jadi, model dapat memprediksi gambar yang diinput ialah gambar kucing
